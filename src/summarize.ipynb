```python
import numpy as np
import pandas as pd
import os,sys
```

```python
def generate_summary_table(ROOT,file_name,ts):
    all_metrics = []
    for rs in [0,1,2]:
        file_dir = os.path.join(ROOT,f'ts_{ts}_rs_{rs}')
        metrics_file = os.path.join(file_dir,file_name)
        metrics_df = pd.read_csv(metrics_file)
        all_metrics.append(metrics_df)
        
    all_metrics_df = pd.concat(all_metrics)
    mean_df = all_metrics_df.groupby('Condition').mean()
    std_df = all_metrics_df.groupby('Condition').std()


    summary_df = mean_df.copy()
    for col in mean_df.columns:
        summary_df[col] = mean_df[col].round(3).astype(str) + '(' + std_df[col].round(3).astype(str) + ')'

    # Reset index for better display
    summary_df.reset_index(inplace=True)
        
    return summary_df 
```
